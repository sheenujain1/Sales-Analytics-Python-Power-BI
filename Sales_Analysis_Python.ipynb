```python
import pandas as pd

data = pd.read_csv(
    "C:/Users/user/Downloads/Sample_Superstore.csv",
    encoding="cp1252"
)
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
print(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
print(data['Ship Date'])
```

```python
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month_name()
print(data['Year'])
print(data['Month'])
```

```python
data = data.drop_duplicates()
```

```python
total_sales = data['Sales'].sum()
print(total_sales)
```

```python
total_profit = data['Profit'].sum()
print(total_profit)
```

```python
sales_region = data.groupby('Region')['Sales'].sum()
print(sales_region)
```

```python
sales_category = data.groupby('Category')['Sales'].sum()
print(sales_category)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_region.plot(kind='bar')

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
monthly_sales = data.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
sns.barplot(x='Category', y='Profit', data=data)

plt.title("Profit by Category")

plt.show()
```
